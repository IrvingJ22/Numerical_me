```python
def Decimal_a_Binario(numero):
    binario = ""

    # Signo '0' si el número es positivo y '1' si el número es negativo
    if numero < 0:
        binario += "1"
        numero = abs(numero)
    else:
        binario += "0"

    # Exponente y Mantisa
    if numero == 0:
        binario += "0" * 31
    else:
        exponent = 127
        normalized_numero = numero

        while normalized_numero >= 2:
            normalized_numero /= 2
            exponent += 1

        binario += format(exponent, "08b")

        mantissa = ""
        fraction = normalized_numero - 1.0

        for _ in range(23):
            fraction *= 2
            bit = int(fraction)
            mantissa += str(bit)
            fraction -= bit

        binario += mantissa
    return binario
```

```python
def Binario_a_Decimal(binario):
    sign = int(binario[0])
    exponent = int(binario[1:9], 2)
    mantissa = 1.0

    for i in range(9, 32):
        mantissa += int(binario[i]) * 2**(-i+8)

    sign_factor = -1 if sign == 1 else 1
    result = sign_factor * mantissa * 2**(exponent-127)
    return result
```

```python
# Punto flotante a representación binaria
print("-- Decimal a representación binaria --")
num = float(input("Introduce un número decimal:")) #numero = -12.345
binario = Decimal_a_Binario(num) #1 bit de signo, 8 de exponente y 23 de mantisa
print("Representación binaria:", binario) 
```

```python
# Binario a representación punto flotante (puede tener un ligero grado de error)
print("-- Binario a representación punto flotante --")
bina = input("Introduce un número binario:") #binario = '11000001010001011000010100011110'
numero = Binario_a_Decimal(bina)
print("Representación número de punto flotante:", numero)
```
