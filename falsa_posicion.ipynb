```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def f(x):
    return (x**3) - 2*x + 1
```

```python
def falsa_posicion(f, a, b, error, iter):
    x = np.linspace(a, b, 1000)
    y = f(x)

    plt.plot(x, y)
    plt.axhline(y=0, color='k')
    plt.axvline(x=0, color='k')
    plt.axvline(x=a, linestyle='--', color='red')
    plt.axvline(x=b, linestyle='--', color='red')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.title('MÃ©todo de la falsa posicion')
    
    if f(a) * f(b) > 0:
        print(f"No hay raiz en el intervalo [{a},{b}]")
        return None
    else:
        for i in range(iter):
            xr = a - ( (f(a) * (b-a)) / (f(b) - f(a)))
            plt.plot(xr, f(xr),"ro")
            
            if abs((xr - a) / xr) < error or f(xr) == 0:
                break
            elif f(xr) * f(a) < 0:
                b = xr
            else:
                a = xr
                
    plt.show() 
    print(f"Raiz en: {xr}")
    print(f"Iteraciones: {i+1}") 
         
    return xr
```
